```python
# Função que cria a variável Nmes com os nomes dos meses do ano

def calcula_nome_mes(mes):
  vmes = mes

  if mes == 1:
    return "Janeiro"
  elif mes == 2:
    return "Fevereiro"
  elif mes == 3:
    return "Março"
  elif mes == 4:
    return "Abril"
  elif mes == 5:
    return "Maio"
  elif mes == 6:
    return "Junho"
  elif mes == 7:
    return "Julho"
  elif mes == 8:
    return "Agosto"
  elif mes == 9:
    return "Setembro"
  elif mes == 10:
    return "Outubro"
  elif mes == 11:
    return "Novembro"
  else:
    return "Dezembro"


# Função que cria a variável num com valores numéricos para as estações do ano

def calcula_estacao_ano(dt):
  vdia = dt.day
  vmes = dt.month

  if vmes in (1,2):
    return "Verão"
  elif vmes == 3:
    if vdia < 21:
      return "Verão"
    else:
      return "Outono"
  elif vmes in (4,5):
      return "Outono"
  elif vmes == 6:
      if vdia < 21:
        return "Outono"
      else:
        return "Inverno"
  elif vmes in (7,8):
      return "Inverno"
  elif vmes == 9:
      if vdia < 21:
        return "Inverno"
      else:
        return "Primavera"
  elif vmes in (10,11):
      return "Primavera"
  elif vmes == 12:
      if vdia < 21:
        return "Primavera"
      else:
        return "Verão"

# Função que cria a variável Nmes com os períodos do dia em função da hora

def calcula_periodo_dia(hora):
    vhora = hora

    if vhora in ("00:00","01:00","02:00","03:00","04:00","05:00","06:00"):
      return "Madrugada"
    elif vhora in ("07:00","08:00","09:00","10:00","11:00","12:00"):
      return "Manhã"
    elif vhora in ("13:00","14:00","15:00","16:00","17:00","18:00"):
      return "Tarde"
    else:
      return "Noite"
```

#### 3.5.4 - Dióxido de Nitrogênio (NO2)

Muitos são os efeitos adversos provocados pela prolongada exposição a altos níveis do poluente dióxido de nitrogênio (NO2). Seus impactos na saúde humana podem variar desde a irritação das vias respiratórias até o agravamento de condições como ataques de asma ou sintomas da doença pulmonar obstrutiva crônica (DPOC), especialmente em pessoas com condições respiratórias pré-existentes. Além disso, não devemos negligenciar os efeitos no meio ambiente, em particular sua contribuição para a má qualidade do ar, com potencial para afetar a flora e fauna, causando danos a ecossistemas aquáticos e terrestres.

É importante lembrar que o NO2 é também um gás de efeito estufa, contribuindo para o aquecimento global. Além disso, desempenha um papel na formação de chuva ácida, que pode prejudicar a vegetação, corpos d'água e solos. Portanto, torna-se imperativo adotar medidas para reduzir as emissões desse poluente, tais como regulamentações mais rígidas para veículos automotores, o uso de tecnologias de controle de poluição e a promoção de meios de transporte mais limpos, como o transporte público e a mobilidade sustentável.
